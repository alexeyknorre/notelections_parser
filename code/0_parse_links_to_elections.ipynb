```python
from bs4 import BeautifulSoup
import csv

with open("/Users/alex/Dropbox/R/notelections_parser/data/notelections_2015_2021_majorit.html") as fp:
    soup = BeautifulSoup(fp, "html.parser")
    
links = soup.find_all('li','list-group-item')
```

```python
lst = []
region = ''
munic_title = ''

for i in links:
    
   region_div = i.span

   if region_div is not None:
        region = region_div.nobr.get_text()
    
   i = i.div
   if i is None:
       continue
    
   munic_title_div = i.div
   if i.div.get_text():
        munic_title = i.div.get_text() 


   link = i.div.next_sibling.a['href'].replace('®', '&reg')
   #base_url = "http://notelections.online"
   #link = base_url+link

   munic_elections_title = i.div.next_sibling.get_text()
    
   lst.append([munic_title, munic_elections_title, link, region])
    
import pandas as pd
df = pd.DataFrame(lst, columns=['munic_title','munic_elections_title','link','region'])
df.to_csv("/Users/alex/Dropbox/R/notelections_parser/data/notelections_majorit_parsed_links.csv", sep=',',index=False)
```
